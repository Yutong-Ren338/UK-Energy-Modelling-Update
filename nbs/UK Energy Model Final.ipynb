## Calculating Average Demand by Energy Source and Sector

```python
def average_on_interval(dataframe, start_year, end_year):
    """
    Calculate the average of the values between start_year and end_year (inclusive) for each combination
    of indices in the given DataFrame. Assumes years are integer column headers.

    Parameters:
    dataframe (pd.DataFrame): The DataFrame with data.
    start_year (int): The starting year.
    end_year (int): The ending year.

    Returns:
    pd.DataFrame: A DataFrame with the average values calculated.
    """

    # Selecting the columns for the specified range of years
    columns_for_average = list(range(start_year, end_year + 1))

    # Checking if the specified columns exist in the DataFrame
    if not all(item in dataframe.columns for item in columns_for_average):
        raise ValueError("One or more specified years are not present in the DataFrame columns.")

    # Calculate the average for the specified range of years
    average_df = dataframe[columns_for_average].mean(axis=1)

    return average_df
```

And a function to plot demand by fuel with sectors on the x-axis. This is useful if we want to visualize how taking the average demand between different years impacts the demand plot.

```python
def plot_stacked_bar(dataframe,x_categories,start_year,end_year):
    """
    Plots the output of the average_on_interval function.
    """
    
    # Selecting the columns for the specified range of years
    columns_for_average = list(range(start_year, end_year + 1))

    # Checking if the specified columns exist in the DataFrame
    if not all(item in dataframe.columns for item in columns_for_average):
        raise ValueError("One or more specified years are not present in the DataFrame columns.")

    # Calculate the average for the specified range of years
    average_df = dataframe[columns_for_average].mean(axis=1)
    

    # Consolidate all energy sources (categories) into a single list
    categories = average_df.index.get_level_values(1).unique()
    
    ## Plot figure ##

    plt.figure(figsize=(12, 6), dpi=200)
    plt.title(f'UK Energy Demand by Sector (TWh), {start_year}-{end_year} Average',fontweight='bold')

    # Create an array for the x-axis positions of each sector's bar
    x = np.arange(len(sectors))

    # Plotting each category's data for each sector and stacking them
    cumulative_bottom = np.zeros(len(sectors))  # Initialize an array to track the cumulative bottom for stacking

    for category in categories:
        values = [average_df.loc[(sector, category),] if (sector, category) in average_df.index else 0 for sector in sectors]
        plt.bar(x, values, bottom=cumulative_bottom, label=category)
        cumulative_bottom += np.array(values)  # Update the cumulative bottom


    # Adding the total above each bar with bold text and a bubble
    for i, total in enumerate(cumulative_bottom):
        plt.text(x[i], total+15, f'{round(total, 2)}', ha='center', va='bottom', 
                 bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.5') 
                 )
        
    # Adding a bubble in the top left with the grand total
    plt.text(-0.5,750, f'Total: {cumulative_bottom.sum():.2f} TWh', ha='left', va='top', 
             bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.5'),
             fontsize=10,fontweight='bold')

    
    # Adjusting x-axis
    plt.xticks(x, sectors)
    plt.ylabel("Energy Demand (TWh)")
    plt.xticks(rotation=45)

    # Adjusting y-axis
    plt.ylim(0,800)
    plt.grid(axis='y')

    # Adding the legend
    plt.legend(title="Energy Sources",fontsize=10)

    plt.show()
```

### Efficiency Gain Assumptions

The most important assumptions to make regarding efficiency gains upon switching from fuel X to electricity are: comparing electric vehicle (EV) energy efficiency with internal-combustion engine vehicle (ICEV) energy efficiency and comparing the energy efficiency of gas heating with the energy efficiency of electric heating (assuming the latter will powered by heat pumps).

- The efficiency comparison ratio is given as *electric vehicle efficency ratio* (EVER) given as:
$$ \text{EVER} = \frac{\left( \frac{1 \text{ gal gas}}{X \text{ "miles per gal gas"}} \right) \left( \frac{115,000 \text{ Btu}}{1 \text{ gal gas}} \right)}{\left( \frac{Y \text{ "kWh per 100 miles"}}{100 \text{ miles}} \right) \left( \frac{3,412 \text{ Btu}}{1 \text{ kWh}} \right)} $$
A [study published in March 2023](https://www.nrel.gov/docs/fy23osti/84631.pdf) by the *US National Renewable Energy Laboratory* found the average EVER in the US to be $4.4$ i.e. the study found that the average EV light vehicle in the US travels $4.4$ times further per unit of energy than the average registered internal combustion engine light vehicle. The figure was found to be higher ($5.1$) for inter-city travel and lower ($3.6$) for highway testing.


- A recent study by the University of Southampton [results summarized here](https://chat.openai.com/c/9ea8c3b1-c182-464b-a178-671eee79042c) showed that heat-pump energy efficiency tends to be c. 4x higher than that of the average gas boiler.

### Error Propagation

Due to the indefinitive nature of assumptions on electrification rates and efficiency gain factors, all demand prediction results will quoted with error bars which will be obtained using standard techniques of error propagation.

Since the predicted demand $P$ is given by:

$$P = D \times E \times \frac{1}{G}$$

where:
- $D$ is the present demand (assumed to be known precisely in this context)
- $E$ is the electrification rate
- $G$ is the efficiency gain factor

Using the standard partial-derivative formula $\Delta P$, can be expressed as:


$$\Delta P = \sqrt{\left(D \times \frac{1}{G} \Delta E\right)^2 + \left(-D \times E \times \frac{1}{G^2} \Delta G\right)^2}$$

## Electricity and Remnant Fossil Fuel Demand Prediction Results

Now output the results for each sector separately.

**Remaining Demand for Fossil Fuel Energy**

The electrification assumptions above pointed to the fact that not all sectors of the economy can be viably electrified. Some, like aviation and shipping, will continue to use fossil fuels in the forseable future. Sizing the energy that will continue to originate from fossil fuels is crucial in order to determine the need for future installations of carbon capture and storage capacities.

```python
def predict_demand(dataframe, start_year, end_year):
    """
    Return electricity demand results with error bars for each sector.
    """
    # Selecting the columns for the specified range of years
    columns_for_average = list(range(start_year, end_year + 1))

    # Checking if the specified columns exist in the DataFrame
    if not all(item in dataframe.columns for item in columns_for_average):
        raise ValueError("One or more specified years are not present in the DataFrame columns.")

    # Calculate the average for the specified range of years
    average_df = dataframe[columns_for_average].mean(axis=1)

    # Dictionary to hold results for each sector
    sector_results = {
        'Agriculture': agriculture_df,
        'Commercial Services': commercialservices_df,
        'Residential': residential_df,
        'Iron & Steel': ironsteel_df,
        'Other Industry sectors': otherindustry_df,
        'Public services': publicservices_df,
        'Transport': transport_df
    }

    # Calculate for each sector and store in the dictionary
    for sector, df in sector_results.items():
        
        df['Present Demand (TWh)'] = average_df[sector].values
        
        df['Predicted Electricity Demand'] = (df['Present Demand (TWh)'] * df['Electrification (0-1)'] * 
            1/df['Efficiency Gain Factor']
        )
        
        df['Error Bars (TWh)'] = df['Present Demand (TWh)'] * (
            ((df['∆(Electrification)'] / df['Efficiency Gain Factor']) ** 2 +
             (df['Electrification (0-1)'] * df['∆(EGF)'] / df['Efficiency Gain Factor'] ** 2) ** 2) ** 0.5
        )
        
        df['Remaining Fossil-Fuel Demand'] = (df['Present Demand (TWh)'] * (1-df['Electrification (0-1)']))
        
        df['Error Bars Fossil Fuels (TWh)'] = (df['Present Demand (TWh)'] * df['∆(Electrification)'])
        
        df['CO2 (mt)'] =  df['Remaining Fossil-Fuel Demand'] * df['kgCO2/kWh'] # conveniently, units work out as mt
        
        

    return sector_results
```

```python
def print_filtered_data(result_dict, columns_to_print, rows_to_print):
    """
    Prints specified columns of DataFrames for given rows in each sector.

    :param result_dict: Dictionary of DataFrames, one for each sector.
    :param columns_to_print: List of column names to be printed.
    :param rows_to_print: List of row labels to filter by in the 'Energy Source' column.
    """
    for sector, df in result_dict.items():
        # Filter the DataFrame
        filtered_df = df[df['Energy Source'].isin(rows_to_print)]

        # Check if the DataFrame contains the specified columns
        if all(column in filtered_df.columns for column in columns_to_print):
            # Print the sector name for clarity
            print("___________")
            print(f"{sector}")
            # Print the specified columns in a table format for the filtered rows
            print(filtered_df[columns_to_print].to_string(index=False))
        else:
            # Print a message if the DataFrame does not contain the specified columns
            print(f"{sector} cannot be displayed. Missing columns.")
```

```python
def compile_filtered_data(result_dict, columns_to_print, rows_to_print):
    """
    Compiles specified columns of DataFrames for given rows in each sector into a single DataFrame.

    :param result_dict: Dictionary of DataFrames, one for each sector.
    :param columns_to_print: List of column names to be included.
    :param rows_to_print: List of row labels to filter by in the 'Energy Source' column.
    :return: A single DataFrame with all the compiled data.
    """
    compiled_data = []

    for sector, df in result_dict.items():
        # Filter the DataFrame
        filtered_df = df[df['Energy Source'].isin(rows_to_print)]

        # Check if the DataFrame contains the specified columns
        if all(column in filtered_df.columns for column in columns_to_print):
            # Add the sector as a column
            filtered_df['Sector'] = sector

            # Append the filtered data to the compiled_data list
            compiled_data.append(filtered_df[columns_to_print + ['Sector']])

    # Concatenate all the dataframes in the list
    final_df = pd.concat(compiled_data)

    # Rearrange columns to put 'Sector' first
    cols = final_df.columns.tolist()
    cols = [cols[-1]] + cols[:-1]  # Move 'Sector' to the beginning
    final_df = final_df[cols]

    return final_df

# Example usage
predicted_demand_df = compile_filtered_data(result_dict, ['Energy Source', 'Predicted Electricity Demand'],
                            ['Electricity', 'Natural Gas', 'Petroleum products', 'Petroleum products (Aviation)',
                             'Petroleum products (Rail)','Petroleum products (Road transport)',
                             'Petroleum products (Shipping)', 'Renewables', 'Solid / manufactured fuels'])
# predicted_demand_df
```

## Tile Average Electricity Year

```python

# Define the date range
start_date = "1980-01-01"
end_date = "2019-12-31"

electricity_demand_daily_average = electricity_demand_daily_average['Average Demand Scaled (TWh)'].to_numpy()


# Sample data, replace with actual values

# Create date range for 40 years including leap years
date_range = pd.date_range(start=start_date, end=end_date)

# Create a function to repeat the daily values for each year, accounting for leap years
def tile_daily_values(daily_values, dates):
    # Initialize an empty array to hold the tiled values
    tiled_values = np.zeros(len(dates))
    
    # Loop through each year in the date range
    current_year = dates[0].year
    leap_year_index = 0
    for i, date in enumerate(dates):
        if date.year != current_year:
            current_year = date.year
            leap_year_index = 0  # Reset index for the new year
        
        # Check if the current year is a leap year
        if pd.Timestamp(current_year, 12, 31).dayofyear == 366:
            # Use the full array including the leap day
            tiled_values[i] = daily_values[leap_year_index]
        else:
            # Use the array excluding the last value (leap day)
            tiled_values[i] = daily_values[leap_year_index if leap_year_index < 365 else leap_year_index - 1]
        
        leap_year_index += 1
        if leap_year_index == 366:  # Reset for the next year
            leap_year_index = 0
    
    return tiled_values

# Tile the electricity demand values across the 40 years
tiled_electricity_demand = tile_daily_values(electricity_demand_daily_average, date_range)

```

```python
# daily renewables output in TWh

def wind_solar_nuclear_output(cf_offshore, cf_onshore, cf_solar, fraction_solar, capacity_renewables, daily_supply_nuclear):
    """Output renewables generation corresponding to the... 
    ...datasets of cf_wind and cf_solar and some critical assumptions"""
    
    S_offshore = capacity_renewables*(1-fraction_solar)*0.7*cf_offshore*24/1000 # TWh 
    
    S_onshore = capacity_renewables*(1-fraction_solar)*0.3*cf_onshore*24/1000
    
    S_solar = capacity_renewables*fraction_solar*cf_solar*24/1000 # TWh
    
    S_daily = S_offshore + S_onshore + S_solar + daily_supply_nuclear
    
    return S_daily
```

```python
def average_supply(renewable_capacity,capacity_nuclear_1,capacity_nuclear_2,capacity_nuclear_3,cf_renewables,cf_nuclear):
    """
    Return average supply from renewables and nuclear for 3 different baseload scenarios.
    """
    # supply in TWh
    S_renewables = renewable_capacity*cf_renewables*24*365/1000
    S_nuclear_1 = capacity_nuclear_1*cf_nuclear*24*365/1000 
    S_nuclear_2 = capacity_nuclear_2*cf_nuclear*24*365/1000 
    S_nuclear_3 = capacity_nuclear_3*cf_nuclear*24*365/1000
    
    return S_renewables+S_nuclear_1, S_renewables+S_nuclear_2, S_renewables+S_nuclear_3
    
    
```

```python
# daily renewables output in TWh

def renewables_output(cf_offshore, cf_onshore, cf_solar, capacity_renewables,capacity_nuclear):
    """Output renewables generation corresponding to the... 
    ...datasets of cf_wind and cf_solar and some critical assumptions"""
    
    S_offshore = cf_offshore*capacity_renewables*share_wind*share_wind_offshore*24/1000 # TWh 
    S_onshore = cf_onshore*capacity_renewables*share_wind*share_wind_onshore*24/1000 # TWh
    S_solar = cf_solar*capacity_renewables*share_solar*24/1000 # TWh
    S_nuclear = load_factor_nuclear*capacity_nuclear*24/1000
    
    S_daily = S_offshore + S_onshore + S_solar + S_nuclear

    
    return S_daily
```

```python
#with efficient use of residual energy, DAC capacity
# Storage Volume assumptions
L_max = 71.0 # TWh
L_0 = L_max
# Electrolyser Power Assumptions
P_ele = 50 # GW
E_max = P_ele*24/1000 # TWh
#DAC capacity
DAC_cap = 27.0 # GW
DAC_max = DAC_cap*24/1000 # TWh
part_renewable_capacities_removal = [250]

for capacity in part_renewable_capacities_removal:
    column_name_S_D = f'S-D(TWh),Ren={capacity}GW'
    column_name_L = f'L (TWh),RC={capacity}GW'
    column_name_R = f'R_ccs (TWh),RC={capacity}GW'
    column_name_RDAC = f'R_dac (TWh),RC={capacity}GW'

    net_supply_df_12[column_name_L] = 0.0  # Initialize column with zeros
    net_supply_df_12[column_name_R] = 0.0 
    net_supply_df_12[column_name_RDAC] = 0.0 

    # Month 1

    #### 1) S-D < 0 (net shortage) ##

    # a) Storage was emptied
    if net_supply_df_12[column_name_S_D][0]<0 and net_supply_df_12[column_name_S_D][0] + L_0*e_out < 0:

        net_supply_df_12[column_name_L].iloc[0] = 0


    # b) Storage prevented a shortage
    if net_supply_df_12[column_name_S_D][0]<0 and net_supply_df_12[column_name_S_D][0] + L_0*e_out > 0:

        net_supply_df_12[column_name_L].iloc[0] = L_0 + net_supply_df_12[column_name_S_D][0]/e_out


    #### 2) S-D > 0 (net surplus and sufficient electrolyser power) ####

    # a) Storage is being filled, but there is not enough energy surplus to fill it entirely
    if net_supply_df_12[column_name_S_D][0] > 0 and net_supply_df_12[column_name_S_D][0]<E_max and net_supply_df_12[column_name_S_D][0]*e_in <= L_max:

        net_supply_df_12[column_name_L].iloc[0] = L_0 + net_supply_df_12[column_name_S_D][0]*e_in

    # b) Storage has been filled fully
    if net_supply_df_12[column_name_S_D][0] > 0 and net_supply_df_12[column_name_S_D][0]<E_max and L_0+net_supply_df_12[column_name_S_D][0]*e_in > L_max:
        print("Storage is full, sufficient ele")

        net_supply_df_12[column_name_L].iloc[0] = L_max
        Used = (L_max - L_0) / e_in #Energy used to fill in the storage
        net_supply_df_12[column_name_R].iloc[0] = net_supply_df_12[column_name_S_D][0] - Used
        if net_supply_df_12[column_name_R][0]<DAC_max:
            net_supply_df_12[column_name_RDAC].iloc[0] = net_supply_df_12[column_name_R][0]        
        else:
            net_supply_df_12[column_name_RDAC].iloc[0] = DAC_max
            
    #### 3) S-D > 0 (net surplus and insufficient electrolyser power) ####

    # a) Storage is being filled, but there is not enough energy surplus to fill it entirely
    if net_supply_df_12[column_name_S_D][0] > 0 and net_supply_df_12[column_name_S_D][0]>=E_max and L_0+E_max*e_in <= L_max:

        net_supply_df_12[column_name_L].iloc[0] = L_0 + E_max*e_in

    # b) Storage has been filled fully
    if net_supply_df_12[column_name_S_D][0] > 0 and net_supply_df_12[column_name_S_D][0]>=E_max and L_0+E_max*e_in > L_max:
        print("Storage is full, insufficient ele")
        net_supply_df_12[column_name_L].iloc[0] = L_max
        Used = (L_max - L_0) / e_in
        if Used <= E_max:
            net_supply_df_12[column_name_R].iloc[0] = net_supply_df_12[column_name_S_D][0] - Used
            if net_supply_df_12[column_name_R][0]<DAC_max:
                net_supply_df_12[column_name_RDAC].iloc[0] = net_supply_df_12[column_name_R][0]
            else:
                net_supply_df_12[column_name_RDAC].iloc[0] = DAC_max
        if Used > E_max:
            net_supply_df_12[column_name_R].iloc[0] = net_supply_df_12[column_name_S_D][0] - E_max
            if net_supply_df_12[column_name_R][0]<DAC_max:
                net_supply_df_12[column_name_RDAC].iloc[0] = net_supply_df_12[column_name_R][0]
            else:
                net_supply_df_12[column_name_RDAC].iloc[0] = DAC_max
            
    # now step forward in time from day 2 to day 14,610
    for i in range(1,len(net_supply_df_12)):

        #### 1) S-D < 0 (net shortage) ####

        # a) Storage was emptied
        if net_supply_df_12[column_name_S_D][i]<=0 and net_supply_df_12[column_name_S_D][i] + net_supply_df_12[column_name_L][i-1]*e_out <= 0:

            net_supply_df_12[column_name_L][i] = 0


        # b) Storage prevented a shortage
        if net_supply_df_12[column_name_S_D][i]<=0 and net_supply_df_12[column_name_S_D][i] + net_supply_df_12[column_name_L][i-1]*e_out > 0:

            net_supply_df_12[column_name_L][i] = net_supply_df_12[column_name_L][i-1] + (net_supply_df_12[column_name_S_D][i])/e_out


        #### 2) S-D > 0 (net surplus and sufficient electrolyser power) ####

        # a) Storage is being filled, but there is not enough energy surplus to fill it entirely
        if net_supply_df_12[column_name_S_D][i] > 0 and net_supply_df_12[column_name_S_D][i]*e_in < L_max and net_supply_df_12[column_name_S_D][i]<=E_max:

            net_supply_df_12[column_name_L][i] = net_supply_df_12[column_name_L][i-1] + net_supply_df_12[column_name_S_D][i]*e_in


        # b) Storage has been filled fully
        if net_supply_df_12[column_name_S_D][i] > 0 and net_supply_df_12[column_name_L][i-1]+net_supply_df_12[column_name_S_D][i]*e_in >= L_max and net_supply_df_12[column_name_S_D][i]<=E_max:

            net_supply_df_12[column_name_L][i] = L_max
            Used = (L_max - net_supply_df_12[column_name_L][i-1]) / e_in #Energy used to fill in the storage
            net_supply_df_12[column_name_R][i] = net_supply_df_12[column_name_S_D][i] - Used
            if net_supply_df_12[column_name_R][i]<=DAC_max:
                net_supply_df_12[column_name_RDAC][i] = net_supply_df_12[column_name_R][i]
            else:
                net_supply_df_12[column_name_RDAC][i] = DAC_max
            # print(f'i={i}, {net_supply_df[column_name_L][i]}, Storage is full, sufficient ele')

        #### 3) S-D > 0 (net surplus and insufficient electrolyser power) ####

        # a) Storage is being filled, but there is not enough energy surplus to fill it entirely
        if net_supply_df_12[column_name_S_D][i] > 0 and net_supply_df_12[column_name_L][i-1]+E_max*e_in < L_max and net_supply_df_12[column_name_S_D][i]>E_max:

            net_supply_df_12[column_name_L][i] = net_supply_df_12[column_name_L][i-1] + E_max*e_in

        # b) Storage has been filled fully
        
        if net_supply_df_12[column_name_S_D][i] > 0 and net_supply_df_12[column_name_L][i-1]+E_max*e_in >= L_max and net_supply_df_12[column_name_S_D][i]>E_max:
            # print(f'i={i}, {net_supply_df[column_name_L][i]}, Storage is full, insufficient ele')

            net_supply_df_12[column_name_L][i] = L_max
            Used = (L_max - net_supply_df_12[column_name_L][i-1]) / e_in #Energy used to fill in the storage
            if Used <= E_max:
                net_supply_df_12[column_name_R][i] = net_supply_df_12[column_name_S_D][i] - Used
                if net_supply_df_12[column_name_R][i]<=DAC_max:
                    net_supply_df_12[column_name_RDAC][i] = net_supply_df_12[column_name_R][i]
                else:
                    net_supply_df_12[column_name_RDAC][i] = DAC_max
            if Used > E_max:
                net_supply_df_12[column_name_R][i] = net_supply_df_12[column_name_S_D][i] - E_max
                if net_supply_df_12[column_name_R][i]<=DAC_max:
                        net_supply_df_12[column_name_RDAC][i] = net_supply_df_12[column_name_R][i]
                else:
                    net_supply_df_12[column_name_RDAC][i] = DAC_max
        
arr_dac = []
for i in net_supply_df_12[column_name_L]:
    arr_dac.append(i)
#minimum storage (must be >20)
print(f'minimum stoarge is {min(arr_dac)}')
#DAC energy
print(f'DAC energy is {net_supply_df_12[f'R_dac (TWh),RC={part_renewable_capacities_removal[0]}GW'].sum()/len(net_supply_df_12)*365}')

dac_capacity = 0
for i in range(0,len(net_supply_df_12)):
    if net_supply_df_12[f'R_dac (TWh),RC={part_renewable_capacities_removal[0]}GW'][i] > 0:
        dac_capacity += 1
    capacity_factor = dac_capacity / len(net_supply_df_12) * 100
print(f'Capacity Factor is {capacity_factor}')
```
